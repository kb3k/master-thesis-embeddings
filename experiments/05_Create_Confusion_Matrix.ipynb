```python
import numpy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cm = np.array([[3,1,3,1,2,2,1,1,0,0,2,3],  # count one for every time chatgpt matches wset human rec
              [1,3,1,1,0,1,2,0,3,1,0,2],   # bertcosim (see 04_BERT_Experiments.ipynb output for raw cosine similarity) top 3
              [0,0,1,0,0,2,1,1,1,0,2,1],   # bertknn top 3
              [0,0,1,3,0,3,2,3,3,3,3,3],   # doc2vecosim top 3 (see 03_Word2Vec_Experiments.ipynb output)
              [0,0,2,1,1,3,3,2,3,1,3,2],   # doc2vecknn top 3
              [0,0,0,3,0,2,2,3,1,1,1,0],   # tfidfcosim (02_TFIDF_Experiments.ipynb)
              [1,0,1,2,0,1,0,2,1,3,0,1],   # tfidfknn
              [0,0,0,0,0,2,1,0,1,1,1,0],   # bowcosim (01_BagOfWord_Experiments.ipynb)
              [1,0,0,0,0,2,0,1,1,3,1,0]])  # bowknn
```

```python
sep='-'
sep_len=40
fig_size=(12, 10)
x_labels = ['Q1','Q2','Q3','Q4','Q5','Q6','Q7','Q8','Q9','Q10','Q11','Q12']
y_labels = ['ChatGPT','BERTcosim','BERTKNN','Doc2VecCosim','Doc2VecKNN','TFIDFcosim','TFIDFknn','BOWcosim','BOWknn']
sns.set(rc={'figure.figsize':fig_size})
sns.heatmap(cm, 
            xticklabels=x_labels,
            yticklabels=y_labels,
            annot=True, 
            cmap = 'YlGnBu')
plt.tight_layout()
plt.title("Vectorization Methods and ChatGPT Performance vs. WSET Certified Humans")
plt.pause(0.05)
```
